# Factorization graphs

Una estructura relevante para estudiar la interacción de los **subgrupos** de un grupo finito es el **factorization graph**. En este grafo, cada vértice representa un **subgrupo propio** del grupo \(G\), y se dibuja una arista entre dos subgrupos \(H\) y \(K\) siempre que su producto \(HK\) sea igual a todo el grupo \(G\). De esta manera, la relación algebraica de factorización se traduce en una estructura gráfica, lo que permite analizar propiedades del grupo mediante herramientas de teoría de grafos, tales como la conectividad, los grados de los vértices, la existencia de cliques o la identificación de subgrupos “centrales” en la factorización.  

Existen algunos puntos clave en la construcción de estos grafos:  

- **Vértices y subgrupos**: solo se incluyen los subgrupos propios, es decir, se excluyen el subgrupo trivial \(\{e\}\) y el grupo completo \(G\), salvo que se quiera una visualización más completa.  

- **Aristas**: una arista conecta dos subgrupos \(H\) y \(K\) si y solo si \(HK = G\), es decir, cualquier elemento del grupo puede expresarse como producto de un elemento de \(H\) y un elemento de \(K\). Esto refleja directamente la capacidad de estos subgrupos de “generar” el grupo mediante su combinación.  

- **Propiedades gráficas**: al igual que en los power graphs, es posible analizar el número de componentes conexas, la distribución de grados de los nodos, la existencia de cliques y otras métricas de teoría de grafos para obtener información sobre la estructura algebraica del grupo y las relaciones entre sus subgrupos.  

El estudio de los factorization graphs proporciona así una perspectiva **complementaria** a los power graphs: mientras estos últimos exploran relaciones de potencia entre elementos individuales, los factorization graphs se enfocan en **cómo los subgrupos se combinan para reconstruir el grupo completo**, revelando patrones de generación y factorización que pueden ser útiles en álgebra, combinatoria y teoría de grupos finitos.

## Conclusión

¿Tienes alguna conjetura sobre la secuencia de grados del grafo de factorización de un grupo cíclico finito?

La secuencia de grados del grafo de factorización de un grupo cíclico $\mathbb{Z}_n$ depende de los divisores de $n$: cada vértice (subgrupo propio) está conectado a los subgrupos cuya orden tiene **mínimo común múltiplo igual a \(n\)**. En particular, si $n$ es primo el grafo está vacío, y si $n$ tiene varios factores primos, los vértices correspondientes a divisores coprimos tendrán mayor grado.
