## Grupos Cíclicos


Los **grupos cíclicos** $C_p$ son grupos finitos generados por un solo elemento $g$, llamado **generador**, de modo que todos sus elementos pueden expresarse como potencias de $g$:

$$
C_p = \{ e, g, g^1, g^2, \dots, g^{p-1} \}
$$

donde $p$ es el orden del grupo y $e$ es el elemento identidad. Estos grupos son abelianos, y todos sus subgrupos son también cíclicos. 

Recordemos que la n-ésima potencia de $g$, notada en caso multiplicativo como $g^n$, se define recursivamente como:

$$
g^n :=
\begin{cases}
e, & \text{si } n = 0,\\[1mm]
g \cdot g^{\,n-1}, & \text{si } n > 0,\\[1mm]
(g^{-1})^{-n}, & \text{si } n < 0,
\end{cases}
$$

donde $e$ es el elemento identidad del grupo.

En Python, la librería **`SymPy`** incluye un módulo de combinatoria de grupos que permite construir grupos cíclicos de manera directa.  
Por ejemplo, para crear un grupo cíclico de orden $p$ se puede usar:

```python
G = CyclicGroup(3)
print(G.print_elements()) 
```

Una estructura relevante para estudiar la relación entre los elementos de un grupo finito es el **power graph**, denotado como $P(G)$ cuando $G$ es el grupo finito en cuestión. En este grafo, cada vértice representa un elemento de $G$ y se dibuja una arista entre dos vértices distintos siempre que uno sea potencia del otro. De esta manera, la relación algebraica de potencia se traduce en una estructura gráfica que permite analizar propiedades del grupo mediante herramientas de teoría de grafos, como la conectividad, los grados de los vértices o la existencia de cliques.

En particular, para los **grupos cíclicos**, donde todos los elementos se generan a partir de un único elemento generador, el power graph $P(G)$ resulta siempre **conexo**, ya que cualquier elemento puede conectarse a cualquier otro a través de la relación de potencia. Esto convierte a los grupos cíclicos en ejemplos ideales para estudiar y visualizar los power graphs de manera sencilla y clara.

Para definir un **power graph** (no dirigido) hacemos uso de **`NetworkX`**, una librería que facilita el manejo y visualización de grafos. Nótese que esto también se puede definir manualmente, construyendo la lista de vértices y agregando aristas según la relación de potencia entre los elementos del grupo.  

Consideremos en primer momento una función que verifique si un elemento es potencia de otro. Esta función será la base para generar las aristas del grafo, conectando cada par de elementos que cumpla con esta relación.

Basta con definir el grafo y conectar los nodos correspondientes a los elementos del grupo. Cada vértice del grafo representa un elemento del grupo finito, y agregamos una arista entre dos vértices si uno de los elementos es potencia del otro.  

De esta manera, el **power graph** refleja directamente la relación algebraica de potencia entre los elementos, y podemos analizar su estructura mediante herramientas gráficas, como la conectividad, los grados de los vértices y la existencia de cliques.

A continuación, graficaremos los **power graphs** de los grupos cíclicos para los primos \(p = 3, 5, 7\). Basta con crear un ciclo que recorra estos valores de \(p\), generando para cada uno los elementos del grupo, construyendo su power graph y visualizándolo. De esta manera, podemos mostrar de manera secuencial y comparativa la estructura de los grafos de potencia de cada grupo cíclico.

```python
orders = [3, 5, 7]
fig, axes = plt.subplots(1, 3, figsize=(10, 6))

for idx, n in enumerate(orders):
    G = CyclicGroup(n)
    graph = PowerGraph(G, directed=True)
    graph.draw(ax=axes[idx],
               title=f"Power Graph de C{n}",
               node_color="skyblue",
               node_size=300)

plt.tight_layout()
plt.show()
```

## Grupos Lineales Generales

Los **grupos lineales generales** $GL_n(p)$ consisten en todas las matrices invertibles de tamaño $n \times n$ sobre el cuerpo finito $\mathbb{F}_p$, donde $p$ es un número primo. La operación del grupo es la multiplicación de matrices módulo $p$, y el elemento identidad es la **matriz identidad** $I_n$.  

Formalmente:

$$
GL_n(p) = \{ A \in M_{n \times n}(\mathbb{F}_p) \mid \det(A) \neq 0 \pmod{p} \}
$$

Algunas propiedades relevantes:  
- $GL_n(p)$ es **no abeliano** para $n \ge 2$.  
- Todos sus elementos son invertibles, y cada matriz tiene un inverso único.  
- La estructura de $GL_n(p)$ es más compleja que la de los grupos cíclicos, lo que hace interesante estudiar sus **power graphs**, ya que la relación de potencia entre matrices puede generar grafos con patrones muy distintos a los de los grupos cíclicos.

```python
class GLGroup:
    def __init__(self, n, p):
        self._n = n
        self._p = p
        self._identity = Matrix.eye(n)
        self._elements = self._generate_elements()
    
    def _generate_elements(self):
        matrices = []
        for entries in product(range(self._p), repeat=self._n*self._n):
            mat = Matrix(self._n, self._n, entries)
            if mat.det() % self._p != 0:  # solo matrices invertibles
                matrices.append(mat.applyfunc(lambda x: x % self._p))
        return matrices
    
    @property
    def identity(self):
        return self._identity
    
    @property
    def n(self):
        """Getter del tamaño de las matrices"""
        return self._n
    
    @property
    def p(self):
        """Getter del módulo"""
        return self._p
    
    def get_elements(self, as_tuples=False):
        if as_tuples:
            return [tuple(tuple(row) for row in mat.tolist()) for mat in self._elements]
        else:
            return self._elements
```

```python
class UltraFastMatrixPowerChecker:
    def __init__(self, n, p):
        self.n = n
        self.p = p
        self.identity_np = np.eye(n, dtype=int)
        self.group_order = self._calculate_gl_order(n, p)
        self.divisors = self._get_divisors(self.group_order)
        self.order_cache = {}  # {(hash): order}
    
    def _calculate_gl_order(self, n, p):
        """Calcula |GL(n,p)| = ∏_{k=0}^{n-1} (pⁿ - pᵏ)"""
        order = 1
        for k in range(n):
            order *= (p**n - p**k)
        return order
    
    def is_power(self, x, y):
        """Verifica si y es potencia de x"""
        # Convertir a arrays numpy
        x_np = np.array(x.tolist(), dtype=int) % self.p
        y_np = np.array(y.tolist(), dtype=int) % self.p
        
        if np.array_equal(x_np, y_np):
            return True
        if np.array_equal(y_np, self.identity_np):
            return True
        
        # Obtener orden de x
        order_x = self._get_order_np(x_np)
        
        # Buscar potencia solo hasta el orden de x
        current = x_np.copy()
        for k in range(2, order_x + 1):
            current = self._np_matmult_mod(current, x_np, self.p)
            if np.array_equal(current, y_np):
                return True
        
        return False
    
    def _get_order_np(self, matrix_np):
        """Encuentra el orden correctamente usando el teorema de Lagrange"""
        matrix_hash = matrix_np.tobytes()
        if matrix_hash in self.order_cache:
            return self.order_cache[matrix_hash]
        
        # Primero verificar si es la identidad
        if np.array_equal(matrix_np, self.identity_np):
            self.order_cache[matrix_hash] = 1
            return 1
        
        # Usar el teorema de Lagrange: el orden debe ser divisor de group_order
        # Buscar el MÍNIMO d (divisor de group_order) tal que matrix^d = I
        
        # Ordenar divisores de menor a mayor para encontrar el mínimo
        for d in sorted(self.divisors):
            if d == 1:
                continue
            power_d = self._np_matrix_power(matrix_np, d, self.p)
            if np.array_equal(power_d, self.identity_np):
                # Verificar que es el mínimo
                is_minimal = True
                for divisor in self.divisors:
                    if divisor < d and d % divisor == 0:
                        # Verificar si un divisor más pequeño también funciona
                        power_divisor = self._np_matrix_power(matrix_np, divisor, self.p)
                        if np.array_equal(power_divisor, self.identity_np):
                            is_minimal = False
                            break
                
                if is_minimal:
                    self.order_cache[matrix_hash] = d
                    return d
        
        # Fallback: no se encontró orden (no debería pasar para matrices invertibles)
        return self.group_order
    
    def _np_matrix_power(self, matrix_np, exponent, p):
        """Exponenciación matricial con numpy"""
        result = np.eye(self.n, dtype=int)
        base = matrix_np
        exp = exponent
        
        while exp > 0:
            if exp % 2 == 1:
                result = self._np_matmult_mod(result, base, p)
            base = self._np_matmult_mod(base, base, p)
            exp //= 2
        
        return result
    
    def _np_matmult_mod(self, A, B, p):
        """Multiplicación de matrices n×n optimizada"""
        C = np.zeros((self.n, self.n), dtype=int)
        for i in range(self.n):
            for k in range(self.n):
                if A[i, k]:  # Optimización: solo si no es cero
                    for j in range(self.n):
                        C[i, j] = (C[i, j] + A[i, k] * B[k, j]) % p
        return C
    
    def _get_divisors(self, n):
        """Encuentra todos los divisores de n"""
        divisors = set()
        for i in range(1, int(math.sqrt(n)) + 1):
            if n % i == 0:
                divisors.add(i)
                divisors.add(n // i)
        return sorted(divisors)
    
    def get_group_order(self):
        """Devuelve el orden del grupo GL(n,p)"""
        return self.group_order
```

```python
class OptimizedMatrixPowerChecker(UltraFastMatrixPowerChecker):
    def _get_order_np(self, matrix_np):
        """Versión optimizada para encontrar el orden mínimo"""
        matrix_hash = matrix_np.tobytes()
        if matrix_hash in self.order_cache:
            return self.order_cache[matrix_hash]
        
        if np.array_equal(matrix_np, self.identity_np):
            return 1
        
        # Filtrar solo divisores que podrían ser el orden
        possible_orders = [d for d in self.divisors if d > 1]
        
        # Buscar el mínimo orden
        for d in possible_orders:
            power_d = self._np_matrix_power(matrix_np, d, self.p)
            if np.array_equal(power_d, self.identity_np):
                # Verificar que es mínimo probando sus divisores primos
                if self._is_minimal_order(matrix_np, d):
                    self.order_cache[matrix_hash] = d
                    return d
        
        return self.group_order
    
    def _is_minimal_order(self, matrix_np, candidate_order):
        """Verifica si candidate_order es realmente el orden mínimo"""
        # Descomponer candidate_order en factores primos
        factors = self._prime_factors(candidate_order)
        
        # Para cada factor primo, verificar si order/factor también funciona
        for factor in set(factors):
            test_order = candidate_order // factor
            power_test = self._np_matrix_power(matrix_np, test_order, self.p)
            if np.array_equal(power_test, self.identity_np):
                return False
        return True
    
    def _prime_factors(self, n):
        """Descompone un número en factores primos"""
        factors = []
        d = 2
        while d * d <= n:
            while n % d == 0:
                factors.append(d)
                n //= d
            d += 1
        if n > 1:
            factors.append(n)
        return factors
```

```python
def create_power_graph(GL):
    """
    Crea el grafo de potencias de un grupo GL(n,p) dado un objeto GLGroup.

    Parámetros
    ----------
    GL : GLGroup
        Objeto que contiene las matrices invertibles n×n sobre GF(p).
    
    Devuelve
    -------
    G : networkx.Graph
        Grafo de potencias.
    labels : dict
        Diccionario de etiquetas legibles para cada nodo.
    """
    # Usar getters de GLGroup
    n, p = GL.n, GL.p
    elements = GL.get_elements()
    
    # Inicializar checker con getters
    checker = UltraFastMatrixPowerChecker(n, p)
    
    # Convertir matrices a numpy para el checker
    elements_np = [np.array(mat.tolist(), dtype=int) % p for mat in elements]
    
    # Inicializar grafo
    G = nx.Graph()
    G.add_nodes_from(GL.get_elements(as_tuples=True))
    
    # Construir aristas según el checker
    for i, x_np in enumerate(elements_np):
        for j, y_np in enumerate(elements_np[i+1:]):
            if checker.is_power(x_np, y_np) or checker.is_power(y_np, x_np):
                G.add_edge(tuple(map(tuple, elements[i].tolist())),
                           tuple(map(tuple, elements[i + j + 1].tolist())))
    
    # Crear etiquetas legibles 2x2 sin paréntesis
    labels = {}
    for mat in elements:
        rows = mat.tolist()
        label = "\n".join(" ".join(str(x) for x in row) for row in rows)
        labels[tuple(map(tuple, rows))] = label
    
    return G, labels
```

```python
GL = GLGroup(2, 2) 

G_GL, labels_GL = create_power_graph(GL)

# Visualización
plt.figure(figsize=(6,6))
nx.draw_circular(G_GL, labels=labels_GL, node_color='skyblue', node_size=500, font_size=10)
plt.title(f"Power Graph de GL({GL.n}, {GL.p})")
plt.show()
```
